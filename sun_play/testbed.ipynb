```python
import sys
caffe_root = '../'
sys.path.insert(0, caffe_root + 'python')
import caffe
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
MODEL_FILE = '../models/logisticregression_deploy.prototxt'
PRETRAINED = 'sunsnapshottest_10000_total_iter_10000.caffemodel'
SINGLE_IMAGE_FILE = '/Users/mprat/Documents/SUN/images/b/bedroom/sun_adjbepzubrcejbye.jpg'
```

```python
net = caffe.Classifier(MODEL_FILE, PRETRAINED,image_dims=(227,227))
net.set_phase_test()
net.set_mode_cpu()
```

```python
# load all the test image files from disk, even though we already have them in a lmdb. SIGH. 
# Oh well, will figure out how to do that in a little bit
test_file = '/Users/mprat/Documents/repos/6867-finalproject/imagesets/Testing_01_with_labels_0_start.txt'
image_root = '/Users/mprat/Documents/SUN/images'
test_imgs_and_labels = []
with open(test_file) as f:
    line = f.readline()
    while line:
        line_split = line.split()    
        test_imgs_and_labels.append((image_root + line_split[0], line_split[-1]))
        line = f.readline()
```

```python
input_image = caffe.io.load_image(SINGLE_IMAGE_FILE)
plt.imshow(input_image)
```

```python
confusion_matrix = np.zeros((397, 397))
for (file_path, label) in test_imgs_and_labels:
    input_image = caffe.io.load_image(file_path)
    prediction = net.predict([input_image])  # predict takes any number of images, and formats them for the Caffe net automatically
#     print 'prediction shape:', prediction[0].shape
#     plt.plot(prediction[0])
    predicted_class = prediction[0].argmax()
    print 'predicted class:', predicted_class
    print 'real label:', label
    confusion_matrix[predicted_class][label] += 1
print confusion_matrix
```

```python
plt.imshow(confusion_matrix, interpolation='nearest')
plt.xticks([])
plt.yticks([])
plt.show()
```
