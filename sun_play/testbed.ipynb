```python
confusion_matrix = np.zeros((397, 397))
```

```python
def update_confusion_matrix():
    out = net.forward()
    prediction_vector = out['loss']
    # print prediction_vector.shape
    prediction_vector = out['loss']
    predicted_class = prediction_vector.argmax(axis=1).flatten()
#     print "predicted class:", predicted_class
    actual_class = caffe.io.blobproto_to_array(net.blobs['label']).flatten().astype(int)
#     print "actual class:", actual_class
    for index in range(len(predicted_class)):
        confusion_matrix[predicted_class[index]][actual_class[index]] += 1
```

```python
for i in range(397):
    print "iteration:", i
    update_confusion_matrix()
```

```python
filename = 'confusion_matrix_logistic_regression_14_percent_accuracy'
# np.save(filename, confusion_matrix)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
cax = ax.matshow(confusion_matrix)
plt.colorbar(cax)
ax.set_xticks([])
ax.set_yticks([])
plt.show()
# plt.savefig(filename + '.pdf')
```

```python
arg_sorts = confusion_matrix.argmax(axis=0).argsort()
reconstructed_mat = np.zeros((397,397))
for row_index in range(397):
    reconstructed_mat[row_index] = confusion_matrix[row_index][arg_sorts]
plt.matshow(reconstructed_mat[arg_sorts])
```
